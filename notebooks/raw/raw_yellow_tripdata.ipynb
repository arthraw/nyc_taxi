```python
import pyspark.sql.functions as F 

df = (
        spark.read.format("parquet")
        .option("inferSchema", True)
        .load("/Volumes/nyc_taxi/transient/transient_vol/yellow_trip/")
        .withColumn("_ingest_at", F.current_timestamp())
        .withColumn("tpep_pickup_date", F.col("tpep_pickup_datetime").cast("date"))
    )
```

```python
(
    df.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .clusterBy("tpep_pickup_date")
    .saveAsTable("nyc_taxi.raw.yellow_tripdata")
)
```
