```python
import pyspark.sql.functions as F 

df = (
        spark.read.format("csv")
        .option("inferSchema", True)
        .option("header", "true")
        .load("/Volumes/nyc_taxi/transient/transient_vol/taxi_zone/")
        .withColumn("_ingest_at", F.current_timestamp())
    )
```

```python
(
    df.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable("nyc_taxi.raw.raw_taxi_zone")
)
```
